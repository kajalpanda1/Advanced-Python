```
import time
```

```
from threading import Thread as t
```

```
count=50000000
```

```
def countdown(n):
  while n>0:
    n-=1
```

```
start=time.time()
```

```
countdown(count)
```

```
end=time.time()
```

```
end-start
```

```
t1=t(target=countdown,args=(count//2,))
t2=t(target=countdown,args=(count//2,))
start=time.time()
t1.start()
t2.start()
t1.join()
t2.join()
end=time.time()
```

```
end-start
```

```
import threading
```

```
cdac=0
lock=threading.Lock()
```

```
lock.acquire()
cdac=cdac+1
```

```
lock.acquire()#waiting for itself to release the lock so deadlock situation
cdac=cdac+2
lock.release()
```

```
print(cdac)
```

```
cdac=0
lock=threading.RLock()
```

```
lock.acquire()
cdac=cdac+1
```

```
lock.acquire()#RLock deals with the deadlock situation
cdac=cdac+2
lock.release()
lock.release()
```

```
print(cdac)
```

```
from multiprocessing import Pool
```

```
start=time.time()
pool=Pool()
a=pool.apply_async(test,args=('Sample1.txt',))
b=pool.apply_async(test,args=('Sample2.txt',))
a.wait()
b.wait()
end=time.time()
```
