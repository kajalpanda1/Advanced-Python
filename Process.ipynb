```
#importing multiprocessing module
import multiprocessing
#cube function
def print_cube(n):
    print("Cube: {}".format(n*n*n))
#square function    
def print_square(n):
    print("Square: {}".format(n*n))
    
if __name__ == "__main__":
#creating processes    
    p1=multiprocessing.Process(target=print_square, args=(10,))
    p2=multiprocessing.Process(target=print_cube, args=(10,))
#starting processes
    p1.start()
    p2.start() 
#halting processes 
    p1.join()
    p2.join()
    
    print('Done!')
```

```
def print_func(continent='Asia'):
    print('Continent name:',continent)
if __name__=='__main__':
    names=['America', 'Europe', 'Africa']
    procs=[]
    p=multiprocessing.Process(target=print_func)
    procs.append(p)
    p.start()
#instantiating process with arguments
    for n in names:
        p=multiprocessing.Process(target=print_func,args=(n,))
        procs.append(p)
        p.start()
#complete the processes
    for p in procs:
        p.join()
```

```
from multiprocessing import Process
#global empty list
result=[]
#square function of given list
def square_list(mylist):
    global result
    for n in mylist:
        result.append(n*n)
    print('p1:{}'.format(result))
if __name__=="__main__":
    mylist=[1,2,3,4]
#creating new process
    p1=Process(target=square_list, args=(mylist,))
#starting process
    p1.start()
#waiting until process finished
    p1.join()
    print('in main:{}'.format(result))
```

```
#function to square alist
def square_list(mylist,result,square_sum):
#append squares of mylist to result
    for idx, num in enumerate(mylist):
        result[idx]=num*num
#sum of the result values
    square_sum.value=sum(result)
    print('p1:{}'.format(result[:]))
    print('p1:{}'.format(square_sum.value))
if __name__=="main":
    mylist=[1,2,3,4]
#creating array of int type with 4 integer values space
    result=multiprocessing.Array('i',4)#'d' for floating point values
#creating value of int type
    square_sum=multiprocessing.Value('i')
    p1=multiprocessing.Process(target=square_list, args=(mylist,result,square_sum))
    p1.start()
    p1.join()
    print("in main:{}".format(result[:]))
    print("in main:{}".format(square_sum.value))
```

```
#using Manager class
import multiprocessing as m
#function to print records in records list
def print_record(records):
  for r in records:
    print('name:',r[0],r[1])
#function to add a new record to records list
def insert_record(r,records):
  records.append(r)
  print('New')

if __name__=="__main__":
  with m.Manager() as manager:
#creating list in server process memory
    records=manager.list([('sam',10),('adam',9),('kevin',9)])
    new_record=('jeff',8)
    p1=m.Process(target=insert_record,args=(new_record,records))
    p2=m.Process(target=print_record,args=(records,))
#running p1 to insert record
    p1.start()
    p1.join()
#running p2 to print records
    p2.start()
    p2.join()
```
