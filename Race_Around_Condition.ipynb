Transaction withdraw deposit: variable updation problem in critical section

```
import multiprocessing as m
```

```
def withdraw(bal):
  for _ in range(10000):
    bal.value=bal.value-1#critical section
```

```
def deposit(bal):
  for _ in range(10000):
    bal.value=bal.value+1#critical section
```

```
def trans():
  bal=m.Value('i',100)
  p1=m.Process(target=withdraw,args=(bal,))
  p2=m.Process(target=deposit,args=(bal,))
  p1.start()
  p2.start()
  p1.join()
  p2.join()
  print(bal.value)
```

```
for _ in range(10):
  trans()
```

Using lock mechanism to overcome race around condition while accessing critical section

```
def withdraw1(bal,lock):
  for _ in range(10000):
    lock.acquire()
    bal.value=bal.value-1
    lock.release()
```

```
def deposit1(bal,lock):
  for _ in range(10000):
    lock.acquire()
    bal.value=bal.value+1
    lock.release()
```

```
def trans1():
  bal=m.Value('i',100)
  lock=m.Lock()
  p1=m.Process(target=withdraw1,args=(bal,lock))
  p2=m.Process(target=deposit1,args=(bal,lock))
  p1.start()
  p2.start()
  p1.join()
  p2.join()
  print(bal.value)
```

```
for _ in range(10):
  trans1()
```
