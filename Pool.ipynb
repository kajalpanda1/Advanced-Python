```
#importing Pool class from multiprocessing module
from multiprocessing import Pool
```

```
import time
```

```
work=(['A',5],['B',2],['C',1],['D',3])
```

```
#function to create workers
def work_log(work_data):
  print(work_data[0],work_data[1])
  time.sleep(int(work_data[1]))
  print(work_data[0])
```

```
#function to map work_log to pool
def pool_handler():
  p=Pool(2)
  p.map(work_log,work)
```

```
pool_handler()
```

```
from multiprocessing import Pool,TimeoutError
```

```
import time
import os
```

```
def f(x):
  return x*x
```

```
with Pool(processes=4) as p:
  print(p.map(f,range(10)))
  for i in p.imap_unordered(f,range(10)):
    print(i)
  res=p.apply_async(f,(20,))
  print(res.get(timeout=1))
  res=p.apply_async(os.getpid,())
  print(res.get(timeout=1))
  multiple_res=[p.apply_async(os.getpid,()) for i in range(4)]
  print([res.get(timeout=1) for res in multiple_res])
  res=p.apply_async(time.sleep,(10,))
  try:
    print(res.get(timeout=1))
  except TimeoutError:
    print('lack patience..timeouterror!')
  print('pool available for work')
print('pool closed.no longer available for work')
```
