```
#importing multiprocessing module
import multiprocessing as m
```

```
#function to send msgs from one end of pipe
def sender(con,msgs):
  for m in msgs:
    con.send(m)
    print(m)
  con.close()
```

```
#function to receive msgs from other end of pipe
def receiver(con):
  while 1:
    m=con.recv()
    if m=='END':
      break
    print(m)
```

```
#msgs to be sent
msgs=['hello','hey','hi','END']
#creating a pipe
parent_con,child_con=m.Pipe()
#creating new processes
p1=m.Process(target=sender, args=(parent_con,msgs))
p2=m.Process(target=receiver, args=(child_con,))
#starting processes
p1.start()
p2.start()
#halting processes
p1.join()
p2.join()
```
