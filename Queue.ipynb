```
#importing Queue class of multiprocessing module
from multiprocessing import Queue
```

```
#creating colors list
colors=['red','green','blue','black']
```

```
count=1
```

```
#creating Queue object
q=Queue()
```

```
print('push items to q')
```

```
for c in colors:
  print('item no:',count,':',c)
  q.put(c)#put(): push items to queue
  count+=1
```

```
print('pop items')
```

```
count=0
```

```
#retrieving items from queue by get()
while not q.empty():
  print('item no:',count,':',q.get())
  count+=1
```

```
import multiprocessing as m
```

```
#function to put() square of list in queue
def sq_list(mylist,q):
  for n in mylist:
    q.put(n*n)
```

```
#get() elements form queue
def print_q(q):
  print('queue elements:')
  while not q.empty():
    print(q.get())
  print('queue not empty')
```

```
mylist=[1,2,3,4]
q=m.Queue()
p1=m.Process(target=sq_list,args=(mylist,q))
p2=m.Process(target=print_q,args=(q,))
p1.start()
p1.join()
p2.start()
p2.join()
```
